```python
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea']=df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
zero_values=['Bedroom2','Bathroom','Car', 'Propertycount','Distance']
df[zero_values]=df[zero_values].fillna(0)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```
