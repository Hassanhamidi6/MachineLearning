```python
import pandas as pd 
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(['Passengerid', 'sibsp','zero','zero.1','zero.2','zero.3','zero.4','zero.5','zero.6',
         'zero.7','zero.8','Parch','zero.9','zero.10','zero.11','zero.12','zero.13',
         'zero.14','zero.15','zero.16','Embarked','zero.17','zero.18','Embarked','Pclass'],axis='columns')
```

```python
inputs=df.drop(['Survived','Embarked'], axis="columns")
target=df['Survived'] 
```

```python
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(inputs, target, test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.naive_bayes import GaussianNB
Model= GaussianNB()
```

```python
Model.fit(X_train, Y_train)
```

```python
Model.score(X_test, Y_test)
```
