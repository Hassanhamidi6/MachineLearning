<h1>LinearRegression()</h1>

```python
# The formula of linear regression is Y=mx+b  
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model 
```

```python
df=pd.DataFrame({
    "area":[2600, 3000, 3200, 3600, 4000],
    "price":[550000, 565000, 610000, 680000, 725000]
})
df
```

```python
plt.xlabel('area(sqft)')
plt.ylabel('price(US)')
plt.scatter(df.area, df.price , marker='+')
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.predict([[5000]])
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area, df.price , color='red' ,marker='+')
plt.plot(df.area,reg.predict(df[['area']]), color='blue')
```

```python
reg.coef_  #  ---> coefficiant 
```

```python
reg.intercept_
```

```python
# Y= mx+b  (Price= m*area+b )
135.78767123*5000+180616.43835616432 # In this we'll get the value which we calculated using reg.predict(5000) 
```
